Many of columns consists of 0 value in it that can't be possble. eg- pregnancies,plasma glucose,blood pressure, triceps skin thickness ,insulin,bmi,diabeties pedigree and age can't be zero.

```python
#Replacing all 0 values in dataset by NaN .
# nans value dfa 
nans = dfa[dfa>0]
nans.head(10)
```

```python
#Check for null values 
nans.isnull().sum()
```

```python
#Checking number of Nan values by each columns ...

print("pregnancies NaN values = ",nans['pregnancies'].isnull().sum())
print("plasma glucose NaN values = ",nans['plasma glucose'].isnull().sum())
print("blood pressure NaN values = ",nans['blood pressure'].isnull().sum())
print("triceps skin thickness NaN values = ",nans['triceps skin thickness'].isnull().sum())
print("insulin NaN values = ",nans['insulin'].isnull().sum())
print("bmi NaN values = ",nans['bmi'].isnull().sum())
print("diabetes pedigree NaN values = ",nans['diabetes pedigree'].isnull().sum())
print("age NaN values = ",nans['age'].isnull().sum())
```


# Now, we have to fill mean values in place of NaN values.

```python
#Mean of each coloums---
print("Mean of insulin = ",dfa['insulin'].mean())
print("Mean of blood pressure = ",dfa['blood pressure'].mean())
print("Mean of triceps skin thickness = ",dfa['triceps skin thickness'].mean())
print("Mean of bmi = ",dfa['bmi'].mean())
print("Mean of pregnancies = ",dfa['pregnancies'].mean())
print("Mean of plasma glucose = ",dfa['plasma glucose'].mean())
```

```python
#replacing nan values by mean value of insulin in insulin coloumn
ins = nans['insulin'].fillna(value = dfa['insulin'].mean()) 
#drop insuline coloumn
dfa.drop('insulin', axis=1) 
# adding a new updated insulin coloum to the data
dfa['insulin'] = ins
dfa.head(5)
```

```python
#replacing nan values by mean value of blood pressure in blood pressure in coloumn
bp = nans['blood pressure'].fillna(value= dfa['blood pressure'].mean()) 
#drop blood pressure coloumn
dfa.drop('blood pressure',axis=1)
# adding a new updated blood pressure coloum to the data
dfa['blood pressure'] = bp
dfa.head(5)
```

```python
#replacing nan values by mean value of triceps skin thickness in triceps skin thickness in coloumn
tst = nans['triceps skin thickness'].fillna(value=dfa['triceps skin thickness'].mean())
#drop triceps skin thickness coloumn
dfa.drop('triceps skin thickness', axis=1)
#adding a new updated triceps skin thickness coloum to the data
dfa['triceps skin thickness'] = tst
dfa.head(5)
```

```python
#replacing nan values by mean value of bmi in bmi in coloumn
bm = nans['bmi'].fillna(value=dfa['bmi'].mean())
#drop bmi coloumn
dfa.drop('bmi', axis=1)
#adding a new updated bmi coloum to the data
dfa['bmi'] = bm
dfa.head(5)
```

```python
#replacing nan values by mean value of plasma glucose in plasma glucose in coloumn
pg = nans['plasma glucose'].fillna(value=dfa['plasma glucose'].mean())
#drop pregnancies coloumn
dfa.drop('plasma glucose', axis=1)
#adding a new updated pregnancies coloum to the data
dfa['plasma glucose'] = pg
dfa.head(5)
```

```python
#replacing nan values by mean value of pregnancies in pregnancies in coloumn
preg = nans['pregnancies'].fillna(value=dfa['pregnancies'].mean())
#drop pregnancies coloumn
dfa.drop('pregnancies', axis=1)
#adding a new updated pregnancies coloum to the data
dfa['pregnancies'] = preg
dfa.head(5)
```
